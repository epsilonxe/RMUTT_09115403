```python
import matplotlib.pyplot as plt
import math
import random
```

$f(x) = x^3$

```python
def f(x):
    y = x ** 3
    return y
```

```python
xmin = -30
xmax = 30
x_step = 0.01

X = []
x = xmin
while x <= xmax:
    X.append(x)
    x = x + x_step

# print(X)
```

```python
Y = []
for x in X:
    y = f(x)
    Y.append(y)
    
plt.plot(X, Y)
Y = []
for x in X:
    y = f(x)
    Y.append(y)
    
plt.plot(X, Y)
```

```python
Z = []
for x in X:
    z = random.randint(1, 20000) * math.cos(x)
    Z.append(z)

plt.plot(X, Z, color='red')
```

```python
plt.plot(X, Y, color='blue')
plt.plot(X, Z, color='red')
```

```python
fig, ax = plt.subplots()
ax.plot(X, Y, color='blue', label='$y = x^3$')
ax.plot(X, Z, color='red', label='$y = rand(10000,2000) * \cos(x) $')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
```
